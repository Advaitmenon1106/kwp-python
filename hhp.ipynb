```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
import seaborn as sns
```

```python
lp2 = pd.read_csv('/home/advait/Desktop/lp2_colgate_dc.csv')
```

```python
print(lp2.head())
```

```python
smf.ols('DENTI_COLGATE_VEDSHAKTI~LP2_market', data=lp2).fit().summary()
```

```python
lp2_ind = lp2.drop(columns=['LP2_market']).columns
betas = np.empty(len(lp2_ind))
```

```python
for i in range (0, len(lp2_ind)):
    formula = ''
    formula = formula+lp2_ind[i]+'~LP2_market'
    model = smf.ols(formula, data=lp2).fit()
    betas[i] = model.params[1]
```

```python
betaVal = pd.DataFrame()
betaVal['Products'] = lp2_ind
betaVal['Betas'] = betas
```
